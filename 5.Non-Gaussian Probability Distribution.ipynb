In statistics, we have **‚Äúmoment‚Äù**.  
The first moment is **‚ÄúMean‚Äù**, the Second moment is **‚ÄúStandard Deviation‚Äù**, the Third moment is **‚ÄúSkewness‚Äù** and the Fourth moment is **‚ÄúKurtosis‚Äù**.  
These are the most important moments in statistics.

- Kurtosis meaning is **‚ÄúCurved‚Äù** or **‚ÄúArching‚Äù**  
- In probability distribution, Kurtosis explains the **‚ÄúTAILDNESS‚Äù** of a real-valued random variable.  
- Example:- In 2007, Sachin played 100 matches with an average of 40 runs per match. Similarly, in 2008, he also played 100 matches with an average of 40 runs. According to the mean, both distributions appear the same, with an average of 40 in both years. However, if we plot the Probability Density Function (PDF) for both years, we can observe differences around the tail. While the mean remains consistent in both graphs, the standard deviation differs, indicating a difference in data spread. In 2008, Sachin exhibited greater consistency, with more values clustered around the mean. This suggests that although Sachin did not achieve a century in 2008, he was more consistent in his performance around the average.  
- Differences in distribution can be assessed based on several factors, including **‚ÄòMean‚Äô**, **‚ÄòStandard Deviation‚Äô**, **‚ÄòSkewness‚Äô**, and **‚ÄòKurtosis‚Äô**.  
- If **‚ÄòSkewness‚Äô** differed, we could gain a different perspective. However, if the mean, standard deviation, and skewness are the same, we can identify differences in data distributions based on **Kurtosis**.  
- By examining both data **‚ÄòProbability Density Function (PDF) Tail Fatness‚Äô**, we can determine whether there are any discrepancies.  
- A PDF with **Fatter Tails** indicates a **Higher Density of Outliers**.  
- On that basis in the above example can say that Sachin had outliers in 2008. This means that while Sachin may have been dismissed without scoring in some matches, he also achieved centuries in others, compared to his 2007 performance.  
- The density of outliers, which we describe through the fatness of the tail, is referred to as **Kurtosis**.

**FATNESS of TAILED** means the presence of **OUTLIER**.

## Excess Kurtosis and Its Types

From sample kurtosis, we subtract **3** and that is **excess kurtosis**, where sample kurtosis refers to the **‚ÄòMeasure of Tailedness‚Äô** or extent of outliers in a sample dataset.

**Definition:**  
Excess kurtosis measures how much a distribution's tails differ from a normal distribution.  
It's calculated as:

**Excess Kurtosis = Kurtosis - 3**  
(since normal distribution has kurtosis = 3)

---

### Interpretation

- **Excess kurtosis = 0** ‚Üí Normal distribution (baseline)  
- **Excess kurtosis > 0** ‚Üí Heavier tails than normal (more extreme values)  
- **Excess kurtosis < 0** ‚Üí Lighter tails than normal (fewer extreme values)  

---

## 1. LEPTOKURTIC (Positive Excess Kurtosis)

**Formula:**  
SK - 3 > 0

**Meaning:**  
LEPTO = *"slender"* ‚Üí sharp, slender peak

**Characteristics:**

- **Fatter tails** than normal distribution  
- **Higher, sharper peak** (high peakedness)  
- **More extreme values/outliers** in the distribution  
- More probability mass in tails = more surprises  

**Examples:**  
Stock market crashes, crypto prices, sudden price movements, high-risk mutual funds  

**Risk:**  
**Riskier and more volatile** than normal distribution ‚Äî expect frequent extreme events  

---

## 2. PLATYKURTIC (Negative Excess Kurtosis)

**Formula:**  
SK - 3 < 0

**Meaning:**  
PLATY = *"broad/flat"* ‚Üí flat, broad peak

**Characteristics:**

- **Thinner tails** than normal distribution  
- **Flatter, broader peak** (low peakedness)  
- **Fewer extreme values/outliers**  
- More uniform, stable distribution  

**Examples:**  
Bounded data, uniform-like distributions, stable processes  

**Risk:**  
**Less risky** ‚Äî extreme events are rare  

---

## 3. MESOKURTIC (Zero Excess Kurtosis)

**Formula:**  
SK - 3 = 0

**Meaning:**  
MESO = *"middle"* ‚Üí normal distribution baseline

**Characteristics:**

- **Same tail fatness** as normal distribution  
- **Same peakedness** as normal distribution  
- Moderate risk of extreme values  
- Any normal distribution (regardless of Œº, œÉ) is mesokurtic  

**Examples:**  
Normal distribution, balanced mutual funds, height/weight data  

**Risk:**  
**Moderate, predictable** ‚Äî standard normal behavior

## 3. STATISTICAL TEST

- There are different tests like **‚ÄúThe Anderson Darling Test‚Äù**, and **‚ÄúShapiro-Wilk-Test‚Äù**
- By using this test with the help of **‚ÄúP-VALUE‚Äù** we figure out whether the distribution we have is Normal or not.
- A **P-VALUE less than the significance level (usually 0.05)** suggests that the data is not normal.

## Shapiro Wilk Test

<p align='center'>
<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*JiFgc21h09HsMWU53roRpQ.png" width=300px><br>
<em>different P-value and density plot look</em>
</p>

- The Shapiro-Wilk test is a statistical test used to assess whether a given sample of data comes from a normally distributed population.
- It tests the null hypothesis that the data is normally distributed.
- Suppose the p-value from the test is less than a chosen significance level (often 0.05). In that case, the null hypothesis is rejected.
- Conversely, if the p-value is greater than the significance level, we fail to reject the null hypothesis.
- It can be performed by using the `shapiro()` function from the `scipy.stats` module.

---

## Anderson Darling Test

- It is another statistical test used to assess whether a given dataset follows a particular distribution, such as the **Normal Distribution**
- It is more sensitive to deviations in the tails of the distribution compared to other normality tests like the Shapiro-Wilk Test
- To perform the Anderson-Darling test in Python, we can use the `anderson()` function from the `scipy.stats` module.
- It has two parameters:
  - **critical values**
  - **significance level**
- If the test statistic is less than the critical value at a chosen significance level, we fail to reject the null hypothesis.
- If the test statistic exceeds the critical value, we reject the null hypothesis.

A QQ Plot is a graphical tool used to assess the similarity of the distributions of two sets of data.

- It is particularly useful for determining whether a set of data follows a normal distribution or not.
- Let's say we have two sets of distributions **[X, Y]**
- X = data / random variable  
- Y = theoretical distribution
- If the points fall on a straight line, both distributions are comparable.

---

## Process of Creating QQ PLOT

1. Take theoretical data or distribution (Y)
2. Sort our data X
3. Find quantiles (1st‚Äì100th percentile) for X
4. Sort Y and find quantiles
5. Plot corresponding quantiles of X and Y
6. If all points fall on a straight line, both distributions are same

## QQ Plot on Uniform Distribution

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate a set of random data
x = np.random.uniform(low=0, high=1, size=1000)
plt.hist(x)
```

```python
# Fit a uniform distribution to the data
params = stats.uniform.fit(x)
dist = stats.uniform(loc=params[0], scale=params[1])

# Create a QQ plot
fig = sm.qqplot(x, dist=dist, line='45')

plt.title('QQ Plot of Uniform Distribution with Uniform Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

In probability theory and statistics, a **uniform distribution** is a probability distribution where all outcomes are equally likely within a given range.

## Summary

- This applies to both **discrete** and **continuous** uniform distributions.
- In a discrete uniform distribution, the range consists of **distinct values**.
- In a continuous uniform distribution, the range represents a **continuous interval of values**.

### PDF ‚Äî Uniform Distribution

- Given range $a$ and $b$, the **probability density** is **constant**.
- Outside this range, the probability density is **zero**.
- The probability density value is: $1 / (b - a)$
- This is because the total area under the graph is 1, and the width is $(b - a)$.

**Mathematically:**

- For $a \le x \le b$: $f(x) = \frac{1}{b - a}$
- For $x < a$ or $x > b$: $f(x) = 0$

---

### CDF ‚Äî Uniform Distribution

- For values less than $a$, **$CDF = 0$**
- For values greater than $b$, **$CDF = 1$**
- Between $a$ and $b$, the CDF increases linearly.
- Formula: $F(x) = \frac{x - a}{b - a}$
- Continuous uniform distribution is symmetric like normal distribution, so **skewness = 0**.

## Log-normal Distribution

- This is another **Continuous Probability Distribution**.
- This distribution is very commonly seen while building **Internet Applications**.
- It is a **heavy-tailed continuous probability distribution**.
- The **logarithm of the random variable is normally distributed**.
- This means the distribution is **right-skewed**, which implies the presence of **outliers**.
- **Not all right-skewed distributions are log-normal**.
- A distribution is called **log-normal** only if:
  - The random variable is right-skewed, **and**
  - Taking the **logarithm of the variable results in a normal distribution**.

**Log-normal distribution is denoted as:** $X \sim \text{Log-Normal}(\mu, \sigma)$  
Where $\ln(X) \sim N(\mu, \sigma)$.

  <img src="https://miro.medium.com/v2/resize:fit:750/format:webp/1*3Y-zEMkiJat6R-pWrQf8tQ.png" width="500">

  <em>Log-Normal PDF for different $\sigma$ values</em>
</p>

- The PDF distribution has two parameters ‚Äî **$\mu$ (mu)** and **$\sigma$ (sigma)**.
- In the visualization above, as the **standard deviation ($\sigma$) increases**, the **spread of the distribution increases**.

---

### Where and Why we see Log-Normal Distributions

1. **Social Media Comments:** If you analyze a famous video with millions of views and count the words in each comment, the resulting distribution is typically Log-Normal.
2. **Logic:** It is logical as very few people write long comments, while the majority write short ones.
3. **Engagement Time:** The time people spend on a video or blog follows this distribution; few read the entire blog, while the majority view it at a glance.
4. **Chess Games:** The number of moves in a chess game also follows a Log-Normal pattern.
5. **Income Distribution:** Income follows this distribution as 97-99% of the population is distributed this way.

---

### PDF Equation of Log-Normal

The Log-Normal PDF equation is very similar to the Normal Distribution with two key modifications:

$$f(x | \mu, \sigma) = \frac{1}{x\sigma\sqrt{2\pi}} \exp\left( -\frac{(\ln x - \mu)^2}{2\sigma^2} \right)$$

1. Instead of subtracting the mean from $x$, we subtract the logarithm of $x$ ($\ln x$) from the mean.
2. We multiply the denominator by $x$ (or divide the whole function by $x$) to perform the log transformation.



---

### CDF Equation of Log-Normal

The Cumulative Distribution Function (CDF) describes the probability that the variable takes a value less than or equal to $x$.

$$F(x) = \Phi\left( \frac{\ln x - \mu}{\sigma} \right)$$

  <br>
  <em>CDF ‚Äî Log Normal</em>
</p>

- The center point reflects the mean, and as the standard deviation $\sigma$ increases, the distribution spreads away from the vertical line representing the mean.

---

### Skewness in Log Normal

Since the distribution is right-skewed, the skewness is always positive and depends on $\sigma$:

$$\text{Skewness} = (e^{\sigma^2} + 2)\sqrt{e^{\sigma^2} - 1}$$

### How to check if a random variable is Log-Normally Distributed?

To verify if a random variable follows a Log-Normal Distribution, follow these steps:

1. **Log Transformation:** Transform the random variable $X$ by taking its natural logarithm: $\ln(X)$.
2. **Normal Check:** The transformed variable $\ln(X)$ should ideally follow a **Normal Distribution**: $\ln(X) \sim N(\mu, \sigma)$.
3. **QQ Plot:** Use a **QQ Plot** to assess whether $\ln(X)$ is approximately normal.
   - Plot the transformed variable $\ln(X)$ against a theoretical Normal Distribution.
   - If the points align mostly along the diagonal line, $\ln(X)$ is normally distributed, indicating the original $X$ is **Log-Normal**.

# Pareto Distribution

<img src="https://miro.medium.com/v2/resize:fit:828/format:webp/1*jJ0hBtic9IRubKzSUgNR4g.jpeg" width="350px">

The **Pareto Distribution** is often referred to as the **80:20 Rule** or the **"Power Law."** It was created by **Vilfredo Pareto** and is most commonly seen in Economics and Finance backgrounds.

It is commonly used to model quantities that exhibit **Power Law Behavior**. In fact, the Pareto distribution is a special case of the Power Law.

---

## Power Law
<img src="https://miro.medium.com/v2/resize:fit:1100/format:webp/1*29v8Ad8g-sPTuDlyzKJg8Q.jpeg" width="350px">

The Power Law follows a very famous rule ‚Äî **80:20**. In mathematics, a power law is a functional relationship between two variables, where one variable is proportional to the power of the other.

Specifically, if $y$ and $x$ are two variables related by power law, then the relationship can be written as:
$$Y = K \cdot X^{\alpha}$$
Where:
* **K**: Some constant
* **X**: The power of alpha ($\alpha$)

<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*hlkdn8izu6UWVH4orYFjIQ.jpeg" width="350px">

### Key Characteristics:
* **80:20 Rule**: 80 percent of the population is controlled by 20 percent of people.
* **Wealth Distribution**: 80 percent of the entire wealth is owned by only 20 percent of people, and the rest 80 percent of people have 20 percent income.
* **Parameter**: In a Pareto Distribution, there is only one primary parameter ‚Äî **"ALPHA"**.

---

## Different Alpha Values & Impact
<img src="https://miro.medium.com/v2/resize:fit:1100/format:webp/1*Vz5Q6owr0AGMN7IL6Ekb8w.png" width="350px">

In this graph, we have 3 Pareto distributions. On the Y-axis, we have **Probability Density**.

* **Red Line ($\alpha = 3$):** High Alpha value results in a high height on the Y-axis and a **Thin Tail**.
* **Blue Line ($\alpha = 2$):** Height is a little less than the red line.
* **Green Line ($\alpha = 1$):** Shallow height and a **Fatter Tail**.

**Summary:**
* **HIGH ALPHA VALUE** $\rightarrow$ THIN TAIL
* **LESS ALPHA VALUE** $\rightarrow$ FATTER TAIL

When ALPHA equals 1, variation is minimal. However, when ALPHA tends toward infinity, the graph manifests as a solid black straight line parallel to the Y-axis with no discernible tail.

### Examples:
* **Human settlements:** Many homes are settled in small areas, but few homes develop with large chunks of area.
* **Internet traffic:** 80 percent of data volume comes from 20 percent of files.
* **Stock market:** 80 percent of brokerage comes from 20 percent of people.
* **Note:** If the Alpha value equals 1.16, then the 80:20 rule operates precisely.

---

## Given a Distribution: How to detect its Pareto Distribution?
We can detect if a distribution is Pareto through a **Log-Log Plot** or a **QQ Plot**.

### 1. Log-Log Plot
First, we take the logarithm of all $X$ values and calculate $Y$ accordingly. Then, we take the logarithm of $Y$ values. We plot $\log(X)$ on the horizontal axis and $\log(Y)$ on the vertical axis. If the result is an **inverted line** (sloping downward), it indicates a Pareto Distribution.

<img src="https://miro.medium.com/v2/resize:fit:196/format:webp/1*i6ozO4WV7TFDQsqXHdCFQQ.png" width="350px">

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the Pareto distribution
alpha = 3
xm = 1

# Create an array of x values
x = np.linspace(0.1, 10, 1000)

# Calculate the y values of the Pareto distribution
y = alpha * (xm**alpha) / (x**(alpha+1))

plt.plot(x,y)
```

```python
# Create the log-log plot
plt.plot(np.log(x), np.log(y))

# Add labels and a title
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Pareto distribution')

# Show the plot
plt.show()
```

Inverted line, hence proved Pareto distribution.

### 2. QQ Plot (Quantile-Quantile Plot)
In a QQ Plot, we compare two distributions (X and Y). We select X as a known Pareto Distribution and plot our random variable Y against it. If the data points align along a straight line, Y follows a Pareto Distribution.

**Python Implementation:**

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Define the parameters
alpha = 2
xm = 1

# Generate random data from the Pareto distribution
x = stats.pareto.rvs(b=alpha, scale=xm, size=1000)

plt.hist(x)
plt.show()
```

<img src="https://miro.medium.com/v2/resize:fit:640/format:webp/1*J97ao0UgVQ2YItcnowxafw.png" width="350px">

```python

# Fit a Pareto distribution to the data
params = stats.pareto.fit(x, floc=0)

# Create a Pareto distribution object with the fitted parameters
dist = stats.pareto(b=params[0], scale=params[2])

# Create a QQ plot
fig = sm.qqplot(x, dist=dist, line='45')

# Add titles
plt.title('QQ Plot of Pareto Distribution with Pareto Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

plt.show()
```

This way, we can verify if a distribution follows the Pareto model.
